```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Database connection configuration
db_config = {
    'user': 'root',
    'password': 'SaiNath@987',
    'host': 'localhost',
    'database': 'EmployeeManagementSystemData'
}

# Connect to the MySQL database
conn = mysql.connector.connect(**db_config)

# Query to retrieve department-wise employee count
query = """
    SELECT department_id, COUNT(employee_id) AS employee_count
    FROM employees
    GROUP BY department_id
"""

# Read the data into a pandas DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['employee_count'], labels=df['department_id'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Employees Across Departments')
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Database connection configuration
db_config = {
    'user': 'root',
    'password': 'SaiNath@987',
    'host': 'localhost',
    'database': 'EmployeeManagementSystemData'
}

# Connect to the MySQL database
conn = mysql.connector.connect(**db_config)

# Query to retrieve project and department data
query = """
SELECT d.department_name, COUNT(p.project_id) AS project_count
FROM project p
JOIN departments d ON p.department_id = d.department_id
GROUP BY d.department_name
"""

# Read the data into a pandas DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Generate the histogram
plt.figure(figsize=(12, 8))
plt.bar(df['department_name'], df['project_count'], color='skyblue', edgecolor='black')
plt.xlabel('Department Name')
plt.ylabel('Number of Projects')
plt.title('Project Distribution by Department')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Database connection configuration
db_config = {
    'user': 'root',
    'password': 'SaiNath@987',
    'host': 'localhost',
    'database': 'EmployeeManagementSystemData'
}

# Connect to the MySQL database
conn = mysql.connector.connect(**db_config)

# Query to retrieve role data from project_assignments
query = """
SELECT role, COUNT(assignment_id) AS role_count
FROM project_assignments
GROUP BY role
"""

# Read the data into a pandas DataFrame
df = pd.read_sql(query, conn)

# Close the connection
conn.close()

# Generate the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['role'], df['role_count'], color='skyblue')
plt.xlabel('Role')
plt.ylabel('Number of Assignments')
plt.title('Number of Assignments per Role')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Conclusion

       Employees Table: Stores comprehensive employee information including their name, department, job title, hire date, and salary.
       Departments Table: Manages department details such as name, manager, location, phone number, and budget allocation.
       Project Table: Contains project-related information like name, start and end dates, budget allocation, and the department it belongs to.
       Project_Assignments Table: Links employees to projects, specifying their roles and assignment dates, along with the current status of the assignment.
       Salaries Table: Stores historical salary details of employees, including the salary amount, effective date, end date, and the reason for the salary change.
       Performance_Review Table: Records performance reviews of employees, capturing details such as the review date, rating, comments, and the ID of the reviewer.
       Employee-Department Relationship:Each employee is associated with a single department, facilitating department-wise operations and reporting, managed through the department_id in the Employees table.
       Project-Department Relationship: Projects are linked to departments to indicate which department is responsible for managing and executing the project, established through the department_id in the Project table.
